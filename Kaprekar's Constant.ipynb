```python
import numpy as np

def number():
    # choosing 4 different numbers
    num = np.random.randint(10000)
    return num

def extract(num):
    n1 = num//1000
    n2 = (num - (n1*1000)) // 100
    n3 = (num - (n1*1000) - (n2*100)) //10
    n4 = (num - (n1*1000) - (n2*100) - (n3*10))
    l = [n1,n2,n3,n4]
    return l

def checkforRepeat(l):
    if l[0] == l[1] == l[2] == l[3]:
        return True
    else:
        return False

def order(l):
    # sorting in decending order
    l.sort(reverse=True)
    dlist = l.copy()
    # sorting in acending order
    l.sort()
    alist = l.copy()
    olist = [dlist,alist]
    return olist

def dalist(olist):
    dlist = olist[0]
    alist = olist[1]
    dnum = (dlist[0]*1000) + (dlist[1]*100) + (dlist[2]*10) + (dlist[3])
    anum = (alist[0]*1000) + (alist[1]*100) + (alist[2]*10) + (alist[3])
    danum = [dnum,anum]
    return danum

def minus(danum):
    num = danum[0] - danum[1]
    return num

def kaprekar(value=None):
    """
    Function willdisplay all the steps of Kaprekar's process till the end.
    It will pick a random integer from (1,9999) except the numbers with all the digits same.
    Arrange the digits in decending order to get the largest possible number
    Arrange the digits in ascending order to get the smallest possible number
    Subtract the smaller number from the larger number
    Repeat the process till attaining the Kaprekar's constant (6174)
    """
    if value is not None and isinstance(value,(int,float)):
        if type(value)==float:
            value=int(value)
            print("Floating point numbers are not allowed\nSo proceeding with the integer part of the number....")
        if value>=10000 or value<=0:
            print("Value Should be less than 10000")
            while True:
                num = int(input("Enter an integer from (1,9999):"))
                if num>=10000 or num<=0:
                    continue
                else:
                    break
        else:
            num = value
        print("Given Number is:",num)
    else:
        num = number()
        print("Choosing a number up to 4 digits:",num)
    while True:
        while True:
            l = extract(num)
            c = checkforRepeat(l)
            print("\nChecking if all digits are same....")
            if c == True:
                print("Positive.\nNumbers with all same digits are not applicable.\nMannual restart needed")
                return
            else:
                print("Negative, Processing....")
                break
        olist = order(l)
        danum = dalist(olist)
        print("\nThe number in decending order:",danum[0])
        print("The number in acending order:",danum[1])
        num = minus(danum)
        print("Subracting small from large:",num)
        for k in range(60):
            print("-",end="")
        if num == 6174:
            break
        print("\n\nTaking the subtracted value:",num)
    print("\n\nAt last the number is: 6174\n\nChecking if all digits are same....\nNegative, Processing....\n\nThe number in decending order: 7641\nThe number in acending order: 1467\nSubracting small from large: 6174")
    for k in range(60):
            print("-",end="")
    print("\n\nFrom here on, 6174 will be constant\nHence, Kaprekar's constant attained")
```
